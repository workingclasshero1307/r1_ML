```python
import pandas as pd
# sample data for SME product usage of LMS
data = {
      "Product": ["Wallet", "Loan", "Payment", "Wallet", "Loan"],
      "Users": [1200, 300, 3400, 1300, 280],
      "Revenue": [5000, 10000, 15000, 5500, 9000]
}
df = pd.DataFrame(data)
df.to_csv("sample_data.csv", index=False)
df.head()
```

```python
# read the csv
df = pd.read_csv("sample_data.csv")

# show first few rows
print(df.head())

# basic stats
print(df.describe())
```

```python
# total users per product
users_per_product = df.groupby("Product")["Users"].sum()
print("\ntotal users per product:\n",users_per_product)

# total revenue per user
revenue_per_product = df.groupby("Product")["Revenue"].sum()
print("\ntotal revenue per product:\n",revenue_per_product)
```

```python
for product in df["Product"].unique():
  users = users_per_product[product]
  revenue = revenue_per_product[product]
  print(f"{product} has {users} users and ${revenue} revenue")
```
